```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn import svm
```

```python
#s=svm.SVC(C=1.0, kernel='linear', gamma='scale')
#s=svm.SVC(C=10, kernel='rbf', gamma='scale')
s=svm.SVC(C=1.0, kernel='poly', degree=2)
```

```python
sv=s.fit(xtrain,ytrain)
```

```python
y_predict=sv.predict(xtest)
y_predict
```

```python

from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy=accuracy_score(ytest,y_predict)
accuracy
```

```python
print(classification_report(ytest,y_predict))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
matrix=confusion_matrix(ytest,y_predict)
matrix
```
